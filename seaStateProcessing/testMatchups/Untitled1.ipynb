```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dates = ds.QS_TIME.to_numpy()
```

```python
plt.figure(figsize = (20, 5))
plt.plot(dates, tao)
plt.plot(dates, qs)
```
