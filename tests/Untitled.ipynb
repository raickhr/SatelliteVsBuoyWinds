```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate

def autocorrelation(x):
    """Calculate the autocorrelation of the time series."""
    n = len(x)
    variance = np.var(x)
    x = x - np.mean(x)
    r = np.correlate(x, x, mode='full')[-n:]
    result = r / (variance * np.arange(n, 0, -1))
    return result

def decorrelation_timescale(acf, dt):
    """Calculate the decorrelation timescale from the autocorrelation function."""
    acf_normalized = acf / acf[0]  # Normalize ACF to start at 1
    times = np.arange(len(acf)) * dt
    decorrelation_time = np.trapz(acf_normalized, times)
    return decorrelation_time

# Example usage:
# Generate a sample time series data
np.random.seed(42)
n_points = 1000
dt = 1.0
time_series = np.cumsum(np.random.randn(n_points))

# Calculate the autocorrelation function
acf = autocorrelation(time_series)

# Calculate the decorrelation timescale
decorrelation_time = decorrelation_timescale(acf, dt)
print(f"Decorrelation Timescale: {decorrelation_time:.2f}")

# Plot the autocorrelation function
plt.plot(np.arange(len(acf)) * dt, acf)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function')
plt.grid()
plt.show()
```
