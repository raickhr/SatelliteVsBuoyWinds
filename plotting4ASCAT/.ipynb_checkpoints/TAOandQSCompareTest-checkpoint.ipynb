```python
def getGreatCircDist(lat1, lon1, lat2, lon2):
    earthRad = 6371.0 #km
    dLambda = np.deg2rad(abs(lon2 - lon1))
    phi1 = np.deg2rad(lat1)
    phi2 = np.deg2rad(lat2)
    ## from Wikipedia
    numerator = np.sqrt((cos(phi2)*sin(dLambda))**2 + 
                        (cos(phi1)*sin(phi2) - 
                         sin(phi1)*cos(phi2)*cos(dLambda))**2)
    
    denominator = (sin(phi1)*sin(phi2) + 
                   cos(phi1)*cos(phi2)*cos(dLambda))
    
    dsigma = np.arctan(denominator/numerator)
    dist = earthRad * dsigma
    return dist
```
