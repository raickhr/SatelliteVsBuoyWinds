```python
import sys
sys.path.append("../COARE3p5/COAREalgorithm/Python/COARE3p5")
from coare35vn import *
import pandas as pd


from netCDF4 import Dataset, num2date, date2num
from datetime import datetime, timedelta, date
import os
from glob import glob

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lat = 5
lon = -140

if lat < 0:
    latUnits = 'S'
else:
    latUnits = 'N'

if lon < 0:
    lonUnits = 'W'
else:
    lonUnits = 'E'

lat=abs(lat)
lon=abs(lon)

bFname = f'../downloads/Buoy/extractedGZ/WINDS/T_{lat:02d}{latUnits}_{lon:03d}{lonUnits}_COARE3p5_2001.nc'
test=f'../downloads/QS_data/{lat}{latUnits}_{lon}{lonUnits}_QS_2000_*.nc'
satFname = glob(f'../downloads/QS_data/T_{lat}{latUnits}_{lon}{lonUnits}_QS_2000_*.nc')[0]
```

```python
satFname
```

```python
bouyDS = Dataset(bFname)
satDS = Dataset(satFname)

satWspd = np.array(satDS.variables['wspd'])
satTime = np.array(satDS.variables['time'])
satTimeUnits = satDS.variables['time'].units
cftimes = num2date(satTime, satTimeUnits)
satDateTime = np.array([datetime(dtm.year, dtm.month, dtm.day, dtm.hour, dtm.minute, dtm.second) for dtm in cftimes])
satDir = -np.array(satDS.variables['wdir']) + 360 + 90

bouyWspd = np.array(bouyDS.variables['U10N'])
bouyTime = np.array(bouyDS.variables['time'])
bouyTimeUnits = bouyDS.variables['time'].units
cftimes = num2date(bouyTime, bouyTimeUnits)
bouyDateTime = np.array([datetime(dtm.year, dtm.month, dtm.day, dtm.hour, dtm.minute, dtm.second) for dtm in cftimes])
bouyDir = -np.array(bouyDS.variables['U10_direction'])

satLen = len(satTime)
```

```python
arr = [datetime(2020,1,1),
      datetime(2020,1,2),
      datetime(2020,1,3),
      datetime(2020,1,4),
      datetime(2020,1,5),
      datetime(2020,1,6)]

avg = pd.to_datetime(pd.Series(arr)).mean().to_pydatetime()
avg
```

```python
selDateTime = []
selWspd = []
selWdir = []

atEnd = False

while not atEnd:
    curDate = satDateTime[0]
    mask = abs(satDateTime - curDate) < timedelta(minutes = 15)
    datArr = satDateTime[mask]
    
    selDateTime.append(pd.to_datetime(pd.Series(datArr)).mean().to_pydatetime())
    selWspd.append(np.mean(satWspd[mask]))
    selWdir.append(np.mean(satDir[mask]))
    satDateTime = satDateTime[~mask]
    satWspd = satWspd[~mask]
    satDir = satDir[~mask]

    if sum(~mask) == 0:
        atEnd = True
        
```

```python
lat = 9
lon = -140

if lat < 0:
    latUnits = 'S'
else:
    latUnits = 'N'

if lon < 0:
    lonUnits = 'W'
else:
    lonUnits = 'E'

lat=abs(lat)
lon=abs(lon)


ds = Dataset(f'../downloads/Buoy/extractedGZ/WINDS/T_{lat:02d}{latUnits}_{lon:03d}{lonUnits}_matchup_2000.nc')
#print(ds.variables.keys())
wspdB = np.array(ds.variables['U10N_TAO'])
wdirB = np.array(ds.variables['U10N_dir_TAO'])

wspdS = np.array(ds.variables['U10N_QS'])
wdirS = np.array(ds.variables['U10N_dir_QS'])

bTime = np.array(ds.variables['time_TAO'])
bTimeUnits = ds.variables['time_TAO'].units
cftimes = num2date(bTime, bTimeUnits)
bTime = np.array([datetime(dtm.year, dtm.month, dtm.day, dtm.hour, dtm.minute, dtm.second) for dtm in cftimes])


sTime = np.array(ds.variables['time_QS'])
sTimeUnits = ds.variables['time_QS'].units
cftimes = num2date(sTime, sTimeUnits)
sTime = np.array([datetime(dtm.year, dtm.month, dtm.day, dtm.hour, dtm.minute, dtm.second) for dtm in cftimes])

wspdS[abs(wspdS) > 1000] = float('nan')
wdirS[abs(wdirS) > 1000] = float('nan')

wspdB[abs(wspdB) > 1000] = float('nan')
wdirB[abs(wdirB) > 1000] = float('nan')
ds.close()
```

```python
timeDiff = abs(sTime - bTime)
print(max(timeDiff), min(timeDiff))
```

```python
fig, axes = plt.subplots(nrows=6, ncols =1, figsize=(20,16))
x = sTime
#plt.plot(wind_dateTimeArr1[st1:ed1], angleOrig[st1:ed1], label = 'buoy')


###################################################################
text = 'Actual Data'

ax = axes[0]
ax.plot(x, wspdS, label = 'QS')
ax.plot(x, wspdB, label = 'buoy')
ax.legend()
ax.text(0.5, 0.9, text, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)
ax.grid(visible=True, which='both', axis='both',alpha = 0.5)

###################################################################
text = 'Diff. in Data'
ax = axes[1]

diff = wspdS-wspdB

mean = np.nanmean(diff)
std = np.nanstd(diff)
tstat = f'mean = {mean:.2f} \n std. dev = {std:.2f}'
ax.text(0.2, 0.7, tstat, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)



ax.plot(x, diff, label = 'QS-buoy')
ax.legend()
ax.text(0.5, 0.9, text, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)
ax.grid(visible=True, which='both', axis='both',alpha = 0.5)

###################################################################
text = 'Ratio of Data'

deno = wspdB.copy()
deno[abs(deno)<1e-5] = float('nan')

r = wspdS/deno
ax = axes[2]
ax.plot(x, r, label = 'QS/buoy')
ax.legend()
ax.text(0.5, 0.9, text, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)
ax.grid(visible=True, which='both', axis='both',alpha = 0.5)

###################################################################

text = 'Actual Data : Direction'
ax = axes[3]
ax.plot(x, wdirS, label = 'QS')
ax.plot(x, wdirB, label = 'buoy')
ax.legend()
ax.text(0.5, 0.9, text, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)
ax.grid(visible=True, which='both', axis='both',alpha = 0.5)

###################################################################
ax = axes[4]
text = 'Diff. in Data : Direction'
diff = (wdirS-wdirB)%360
diff[diff > 180] -= 360 
diff[diff < -180] += 360 

mean = np.nanmean(diff)
std = np.nanstd(diff)
tstat = f'mean = {mean:.2f} \n std. dev = {std:.2f}'
ax.text(0.2, 0.7, tstat, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)


ax.plot(x, diff, label = 'QS-buoy')
ax.legend()
ax.text(0.5, 0.9, text, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)
ax.grid(visible=True, which='both', axis='both',alpha = 0.5)

###################################################################

text = 'Ratio of Data : Direction'

deno = wdirB.copy()
deno[abs(deno)<1e-5] = float('nan')

r = wdirS/deno
ax = axes[5]
ax.plot(x, r, label = 'QS/buoy')
ax.legend()
ax.text(0.5, 0.9, text, horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes)
ax.grid(visible=True, which='both', axis='both',alpha = 0.5)

###################################################################

text = f'{lat:d}{latUnits} {lon:d}{lonUnits} 5min-matchup buoy Neutral Winds with QS data 2000'
fig.suptitle(text, y = 0.9)
plt.subplots_adjust(hspace=0.5)

fname = f'images/{lat:02d}{latUnits}_{lon:03d}{lonUnits}_Comp_2000.pdf'
plt.savefig(fname,dpi=70)

```
