```python
ds = Dataset('/home/shikhar.rai/WHOI/SatelliteVsBuoy/downloads/Buoy/mergedTAO_Bouy2000/T0N140W_2000.nc')
```

```python
ds.variables.keys()
```

```python
cdfTime = ds.variables['TIME']
timeUnit = cdfTime.units
timeArr = np.array(cdfTime)
```

```python
cftimes=num2date(timeArr, timeUnit)
dateTimeArr = [datetime(dtm.year, dtm.month, dtm.day, dtm.hour, dtm.minute, dtm.second) for dtm in cftimes]
```

```python
speed = np.array(ds.variables['WSPD'])
```

```python
st = 0
ed = 500
plt.figure(figsize=(20,5))
plt.plot(dateTimeArr[st:ed], speed[st:ed])
```
