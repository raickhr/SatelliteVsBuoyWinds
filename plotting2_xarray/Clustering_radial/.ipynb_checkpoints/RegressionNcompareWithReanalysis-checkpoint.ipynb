```python
df['cos(Direction Difference (QuikSCAT - TAO))'] = np.cos(np.deg2rad(df['Direction Difference (QuikSCAT - TAO)']))
df['sin(Direction Difference (QuikSCAT - TAO))'] = np.sin(np.deg2rad(df['Direction Difference (QuikSCAT - TAO)']))


for timeWindow in range(20,130,10):
    df[f'cos(Direction Difference (QuikSCAT - TAO {timeWindow} min mean))'] = np.cos(np.deg2rad(df[f'Direction Difference (QuikSCAT - TAO {timeWindow} min mean)']))
    df[f'sin(Direction Difference (QuikSCAT - TAO {timeWindow} min mean))'] = np.sin(np.deg2rad(df[f'Direction Difference (QuikSCAT - TAO {timeWindow} min mean)']))
```

```python
timeWindow = 90
# selectX = [f'Speed Difference (QuikSCAT - TAO) {timeWindow} min mean)',
#           f'cos(Direction Difference (QuikSCAT - TAO)) {timeWindow} min mean))',
#           f'sin(Direction Difference (QuikSCAT - TAO)) {timeWindow} min mean))']

selectX = [f'Speed Difference (QuikSCAT - TAO)',# {timeWindow} min mean)',
          f'cos(Direction Difference (QuikSCAT - TAO))',# {timeWindow} min mean))',
          f'sin(Direction Difference (QuikSCAT - TAO))']# {timeWindow} min mean))']


X = df[selectX]

normX = (X - X.mean(axis=0))/ X.std(axis=0)

# kmeans = KMeans(n_clusters=17, random_state=0, n_init="auto", max_iter = 10000).fit(normX)
# df['label'] = kmeans.labels_

dbscan = DBSCAN(eps=0.15, min_samples= 500)
df['label'] = dbscan.fit_predict(normX)

# optics = OPTICS(min_samples=50 ) #cluster_method='dbscan', eps=0.25)
# df['label'] = optics.fit_predict(normX)

```

```python
df['nudge wind direction'] = (ds['nudge_wind_direction']-90)%360
dummy = df['nudge wind direction']
mmask = dummy >180
dummy[mmask] = dummy[mmask] - 360
df['nudge wind direction'] = dummy

df['TAO - nudge speed'] = df['Wind Speed (TAO)'] - ds['nudge_wind_speed']
df['TAO - nudge direction'] = (df['Wind Direction (TAO)'] - df['nudge wind direction'])%360
dummy = df['TAO - nudge direction']

mmask = dummy >180
dummy[mmask] = dummy[mmask] - 360
df['TAO - nudge direction'] = dummy

df['cos(TAO - nudge direction)'] = np.cos(np.deg2rad(df['TAO - nudge direction']))
df['sin(TAO - nudge direction)'] = np.sin(np.deg2rad(df['TAO - nudge direction']))



df['QuikSCAT - nudge speed'] = df['Wind Speed (QuikSCAT)'] - ds['nudge_wind_speed']
df['QuikSCAT - nudge direction'] = (df['Wind Direction (QuikSCAT)'] - df['nudge wind direction'])%360
dummy = df['QuikSCAT - nudge direction']
mmask = dummy >180
dummy[mmask] = dummy[mmask] - 360
df['QuikSCAT - nudge direction'] = dummy

df['cos(QuikSCAT - nudge direction)'] = np.cos(np.deg2rad(df['QuikSCAT - nudge direction']))
df['sin(QuikSCAT - nudge direction)'] = np.sin(np.deg2rad(df['QuikSCAT - nudge direction']))

df['sin(Wind Direction (TAO))'] = np.sin(np.deg2rad(df['Wind Direction (TAO)']))
df['cos(Wind Direction (TAO))'] = np.cos(np.deg2rad(df['Wind Direction (TAO)']))

df['sin(Wind Direction (QuikSCAT))'] = np.sin(np.deg2rad(df['Wind Direction (QuikSCAT)']))
df['cos(Wind Direction (QuikSCAT))'] = np.cos(np.deg2rad(df['Wind Direction (QuikSCAT)']))

df['SST-AIRT'] = df['Sea Surface Temperature (TAO)'] - df['Air Temperature (TAO)']
```

```python
Xheadings = [ 'SST-AIRT',
             'Relative Humidity (TAO)',
             'Neutral Wind Speed at 10m (TAO)',
             'sin(Wind Direction (TAO))', 
             'cos(Wind Direction (TAO))', 
             'sin(Wind Direction (QuikSCAT))',
             'cos(Wind Direction (QuikSCAT))',
             'std. dev. sinWDIR 120min',
             'std. dev. cosWDIR 120min',
             'std. dev. WSPD_10N 120min',
             'cos(TAO - nudge direction)', 
             'sin(TAO - nudge direction)', 
             'cos(QuikSCAT - nudge direction)', 
             'sin(QuikSCAT - nudge direction)',
             'TAO - nudge speed',
             'QuikSCAT - nudge speed']


Yheadings = ['Speed Difference (QuikSCAT - TAO)',
             'cos(Direction Difference (QuikSCAT - TAO))',
             'sin(Direction Difference (QuikSCAT - TAO))']



subDF = df.loc[df['label'].isin([-1])]

X = subDF[Xheadings]
Y = subDF[Yheadings]

X = X.rename(columns=lambda x: x.replace(" ", "_"))
Y = Y.rename(columns=lambda x: x.replace(" ", "_"))




# Generate interaction terms
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_interactions = poly.fit_transform(X)

# Fit regression with interaction terms
model_inter = LinearRegression()
model_inter.fit(X_interactions, Y)

# Extract feature names and coefficients
feature_names = poly.get_feature_names_out(X.columns)
M_inter = model_inter.coef_

# Create DataFrame for interpretation
coef_df_inter = pd.DataFrame(M_inter, columns=feature_names, index=Yheadings)

#tools.display_dataframe_to_user(name="Coefficient Matrix with Interactions", dataframe=coef_df_inter)

from IPython.core.display import display, HTML

# Convert to HTML and display in Jupyter Notebook
display(HTML(coef_df_inter.to_html()))
```
